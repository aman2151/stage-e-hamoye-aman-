```python
# Display the information of ADF test
def show_adfuller(adfull):
  print(f'ADF Statistic: {adfull[0]}')
  print(f'p-value: {adfull[1]}')
  print(f'No. of lags used: {adfull[2]}')
  print(f'No. of observations used : {adfull[3]}')
  print('Critical Values:')
  for k, v in adfull[4].items():
      print(f' {k}: {v}')
```

```python
# jovian.commit(project="stage-e-quiz")
```

```python
from statsmodels.tsa.stattools import adfuller
sys_adf_result = adfuller(daily_df['SysLoad'])
gas_adf_result = adfuller(daily_df['GasPrice'])
```

```python
# Splitting the daily_df into  train and test 
train_data = daily_df.iloc[:2757,:]
test_data = daily_df.iloc[2757:,:]
train_data.shape
```

```python
test_data.shape
```

```python
train_data.reset_index(inplace = True)
train_data.head(5)
```

```python
train_df = train_data.iloc[:,[0,-1]]
train_df = train_df.rename(columns={"FullDate": "ds", "ElecPrice": "y"})
train_df.head()
```

```python
test_data.reset_index(inplace = True)
test_data.head()
test_df = test_data.iloc[:,[0,-1]]
test_df = test_df.rename(columns={"FullDate": "ds", "ElecPrice": "y"})
test_df.head()
```

```python
from fbprophet import Prophet
model = Prophet()
model.fit(train_df)

forecast_val = model.predict(test_df) 
forecast_val.head()
```

```python
import numpy as np

def Metrics(y_true,y_pred):
  y_true,y_pred = np.array(y_true), np.array(y_pred)
  return np.mean(np.abs((y_true - y_pred)/y_true)) *100
```

```python
mape = Metrics(test_df['y'],forecast_val['yhat'])
print("mean absolute percentage error is : ", round(mape,2))
```

```python
from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(test_df['y'],forecast_val['yhat']))
print("mean absolute percentage error is : ", round(rmse,2))
```

```python
model.plot_components(forecast_val)
```

```python
##Setting the regression and prediction features for training
daily_df_new = daily_df.rename(columns = {'FullDate':'ds','Tmax':'add2',
                                           'SysLoad':'add1','GasPrice':'add3','ElecPrice':'y'})
daily_df_new.head() 
```
